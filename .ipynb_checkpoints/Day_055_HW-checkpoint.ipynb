- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
K_init_list=[('7-means',KMeans(n_clusters=7)),
             ('7-means_random',KMeans(n_clusters=7,n_init=1,init='random')),
             ('5-means',KMeans(n_clusters=5)),
             ('5-means_random',KMeans(n_clusters=5,n_init=1,init='random')),
             ('3-means',KMeans(n_clusters=3)),
             ('3-means_random',KMeans(n_clusters=3,n_init=1,init='random')),]
```

```python
num=1
for title,est in K_init_list:
    fig=plt.figure(num,figsize=(4,3))
    ax= Axes3D(fig, rect=[0, 0, .95, 1], elev=20, azim=134)
    est.fit(X)
    ax.scatter(X[:,0],X[:,1],X[:,2],c=est.labels_.astype(np.float),edgecolor='k')
    
    ax.w_xaxis.set_ticklabels(['1st feature'])
    ax.w_yaxis.set_ticklabels(['2nd feature'])
    ax.w_zaxis.set_ticklabels(['3rd feature'])
    
    ax.set_title(title)
    
    ax.dist=12
    num = num + 1
```
