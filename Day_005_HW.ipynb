```python
mean=dummies_app_train.mean()
std=dummies_app_train.std()
mean_std_df=pd.concat([mean,std],axis=1,keys=["Mean","Standard Deviation"])
mean_std_df=mean_std_df.T
```

```python
i=0
fig=plt.figure(figsize=(36,54))


for col in dummies_app_train.columns:
    i+=1
    plt.subplot(25,10,i)
    plt.hist(dummies_app_train[col])
    plt.xlabel(col)
    locs, labels = plt.xticks()
    plt.tick_params(axis='x', which='major', labelsize=6, pad=-6)
    plt.tick_params(axis='y', which='major', labelsize=6)

plt.show()
    
```
